```python
# imports
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('food_data_csv/FOOD-DATA-GROUP1.csv')
data2 = pd.read_csv('food_data_csv/FOOD-DATA-GROUP2.csv')
data3 = pd.read_csv('food_data_csv/FOOD-DATA-GROUP3.csv')
data4 = pd.read_csv('food_data_csv/FOOD-DATA-GROUP4.csv')
data5 = pd.read_csv('food_data_csv/FOOD-DATA-GROUP5.csv')
data1.shape, data2.shape, data3.shape, data4.shape, data5.shape
```

```python
df = pd.concat([data1, data2, data3, data4, data5], ignore_index=True)
df
```

```python
df['sugar per calorie'] = df['Sugars'] / df['Caloric Value']
df
```

```python
def sugar_level(value):
    if value > 0.025:
        return 'red'
    elif 0.015 <= value <= 0.025:
        return 'yellow'
    else:
        return 'green'
```

```python
df['sugar health'] = df['sugar per calorie'].apply(sugar_level)
X = df[['food', 'Caloric Value', 'Sugars', 'sugar per calorie', 'sugar health']].set_index('food')
X
```

```python
X['sugar health'].value_counts()
```

```python
matches = pd.read_csv('food_matches.csv').set_index('menu_item').get(["match"])
matches
```

```python
new_col =[]
for food in matches.get("match"):
    food = food.strip("[]")
    food = food.split(", ")
    food = [name.strip("'") for name in food]
    new_col += [food]

matches = matches.assign(match = new_col)
```

```python
rating_col = []
for food_list in matches.get("match"):
    this_food_ratings =[]
    for food in food_list:
        this_food_ratings += [X.get("sugar health").loc[str(food)]]
    rating_col += [this_food_ratings]

matches = matches.assign(rating = rating_col)
matches
```

```python
new = pd.DataFrame().assign(best_match = [match[0] for match in matches.get('match')]).set_index('best_match')
new = new.merge(df, left_index=True, right_on='food').rename(columns={'food':'best match'}).set_index("best match").drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
new
```

```python
def output_info(filtered_nutrient, df=df):
    df = pd.DataFrame(df)
    df[str(filtered_nutrient) + " per calorie"] = df[filtered_nutrient] / df['Caloric Value']

    #ADJUST LEVELS HERE
    def nutrient_level(value):
        if value > 0.025:
            return 'red'
        elif 0.015 <= value <= 0.025:
            return 'yellow'
        else:
            return 'green'
    df[str(filtered_nutrient) + " health lvl"] = df[str(filtered_nutrient) + " per calorie"].apply(nutrient_level)
    X = df.set_index('food') 
    matches = pd.read_csv('food_matches.csv').set_index('menu_item').get(["match"])
    new_col =[]
    for food in matches.get("match"):
        food = food.strip("[]")
        food = food.split(", ")
        food = [name.strip("'") for name in food]
        new_col += [food]

    matches = matches.assign(match = new_col)
    rating_col = []
    for food_list in matches.get("match"):
        this_food_ratings =[]
        for food in food_list:
            this_food_ratings += [X.get(str(filtered_nutrient) + " health lvl").loc[str(food)]]
        rating_col += [this_food_ratings]

    matches = matches.assign(rating = rating_col)
    return matches

info_df = output_info('Sugars', df=df)
info_df
```

```python
def get_details(rating_df):
    details = pd.DataFrame().assign(best_match = [match[0] for match in rating_df.get('match')]).set_index('best_match')
    details = details.merge(df, left_index=True, right_on='food').rename(columns={'food':'best match'}).set_index("best match").drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
    return (details)
get_details(info_df)
```
